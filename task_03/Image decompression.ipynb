## PCA in Image compression

[blog](https://towardsdatascience.com/image-compression-using-principal-component-analysis-pca-253f26740a9f)

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
img = mpimg.imread('rose.jpg')
```

```python
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
img
```

```python
print('Size of the Original image:',img.shape)
```

```python
x = img.shape[0]
y = img.shape[1]

print(x, y)
```

```python
img_re=np.reshape(img,(x,y*3))
```

```python
print('Reshape image is :',img_re.shape)
```

```python
pca = PCA(100)
```

```python
img_pca=pca.fit_transform(img_re)
```

```python
print('PCA Reshape image is :',img_pca.shape)
```

#### inverse_transform(X)

Transform data back to its original space.

```python
img_inv=pca.inverse_transform(img_pca)
```

```python
img_reshape=np.reshape(img_inv,(x, y,3))
```
