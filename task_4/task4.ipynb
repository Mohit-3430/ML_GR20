```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/tips.csv')
```

```python
df.head()
```

```python
plt.scatter(df['day'], df['tip'])
plt.title('Scatter plot day vs tip')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.scatter(df['day'], df['tip'], c=df['size'], s=df['total_bill'])
plt.title('scatter plot day vs tip')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```
